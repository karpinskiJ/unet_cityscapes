```python
import sys
sys.path.append('../../unet_cityscapes')
import os
from unet_cityscapes.network.UNet import UNet
from unet_cityscapes.dataset.CityscapesDataset import CityscapesDataset
from torch.utils.data import DataLoader
import torch
from matplotlib import pyplot as plt
from torch.utils.tensorboard import SummaryWriter
from datetime import datetime
import torch.nn.functional as F
import torchvision.transforms as T
import cv2
from PIL import Image
import numpy as np
```

```python
dataset_path = "C:\\projects\\unet_cityscapes\\data_512_256_8classes"
shape = (256,512)
samples_cnt = 10
num_classes = 8
batcg_size = 1
loss_fn = torch.nn.CrossEntropyLoss()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
learning_rate = 0.001
```

```python
trainig_dataset = CityscapesDataset(dataset_path,"train",shape= shape,samples_cnt= samples_cnt,num_classes=num_classes)
validation_dataset = CityscapesDataset(dataset_path,"val",shape=shape,samples_cnt= samples_cnt,num_classes=num_classes)
training_loader = DataLoader(trainig_dataset, batch_size=1, shuffle=True)
validation_loader = DataLoader(validation_dataset, batch_size=1, shuffle=True)
```

```python
model = UNet(3, 8).to(device)
model.compile(learning_rate=learning_rate, loss_fn=loss_fn, training_loader=training_loader, validation_loader=training_loader)
```

```python
model.fit(epochs=20)
```
